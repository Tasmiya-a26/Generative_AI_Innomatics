```python
def analyze_engagement(likes_list):
    total_likes = 0
    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like
    # Checking post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    return total_likes, status
# Sample Data
likes = [200, 150, 300, 400]
total, post_status = analyze_engagement(likes)
print("Total Likes:", total)
print("Post Status:", post_status)
```

```python
def rainfall_checker(rainfall_data, required_level):
    total = 0
    # Calculating total rainfall
    for rain in rainfall_data:
        total += rain
    average = total / len(rainfall_data)
    # Checking adequacy
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    return average, status
# Sample Data
rainfall = [70, 80, 75, 65, 70]
required = 70
avg, rain_status = rainfall_checker(rainfall, required)
print("Average Rainfall:", avg)
print("Rainfall Status:", rain_status)
```

```python
# given
def duplicate_detect(usernames):
    unique = set()
    duplicates = set()

    for user in usernames:
        if user in unique:
            duplicates.add(user)   # flagging duplicate
        else:
            unique.add(user)

    # Check if duplicates exist
    if duplicates:
        print("Flagged Accounts:", ", ".join(duplicates))
    else:
        print("Accounts are unique.")

usernames_list = ["ali", "browny", "carol", "ali", "dawood", "browny"]
duplicate_detect(usernames_list)
```

```python
def check_monitor_health(Errors):
    if Errors == 0:
        status = "Healthy"
    elif Errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    return Errors,status

error_list = [7,0,5,3]
for e in error_list:
    count, status = check_monitor_health(e)
    print(f"Error Count:{count}")
    print(f"System Status: {status}")
```
