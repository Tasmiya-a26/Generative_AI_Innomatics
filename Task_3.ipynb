```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# Find the maximum performance score
top_score = max(employees.values())

# Collect all employees who have the top score
eligible_employees = []

for name, score in employees.items():
    if score == top_score:
        eligible_employees.append(name)

# Print the result
print(f"Top Performers Eligible for Bonus: {', '.join(eligible_employees)} (Score: {top_score})")
```

```python
import string
query = "Buy mobile phone buy phone online"
query = query.lower()
query = query.translate(str.maketrans('', '', string.punctuation))
words = query.split()
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1
repeated_words = {word: count for word, count in word_count.items() if count > 1}
print(repeated_words)
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:   # Check if the reading is even
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1
total_emails = len(emails)
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
average_sales = sum(sales) / len(sales)

threshold = average_sales * 1.25   # 25% above average

for index, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {index}: {value}")
```
