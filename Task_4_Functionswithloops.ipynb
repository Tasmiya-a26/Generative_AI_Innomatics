```python
def validate(amount):
    if amount < 50:
        return False
    if amount not in [199, 299, 399, 599]:
        return False
    return True

while True:
    try:
        amt = int(input("Enter recharge amount: "))
    except ValueError:
        print("Please enter a valid number.")
        continue

    if validate(amt):
        print("Recharge successful")
        break
    else:
        print("Invalid amount, try again")
```

```python
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert (Only {quantity} items left)")
        else:
            print(f"{product}: Stock OK ({quantity} items available)")

# Sample inventory dictionary
inventory = {
    "Rice": 40,
    "Sugar": 10,
    "Oil": 25,
    "Salt": 8,
    "Wheat": 15
}

# Function call
check_inventory(inventory)
```

```python
def process_result(marks):
    total = 0
    # Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"

# Sample marks list
student_marks = [60, 45, 70, 55, 40]
# Function call
avg, result = process_result(student_marks)

print(f"Average: {avg}")
print(f"Result: {result}")
```

```python
def calculate_fare(km, peak):
    fare = 50 + (km * 12)
    if peak == "yes":
        fare = fare + (fare * 0.25)
    return fare


while True:
    distance = int(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ")

    print("Total Fare: ₹", calculate_fare(distance, peak_hour))

    retry = input("Retry? (yes/no): ")
    if retry != "yes":
        break
```

```python
def check_eligibility(attendance_list):
    present_count = 0

    # Count number of "P"
    for status in attendance_list:
        if status == "P":
            present_count += 1

    total_days = len(attendance_list)

    if total_days == 0:
        return "No attendance data available"

    percentage = (present_count / total_days) * 100

    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"

# Sample attendance list
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

# Function call
percent, result = check_eligibility(attendance)
print("Attendance Percentage:", round(percent, 2), "%")
print("Status:", result)
```

```python
def check_password(password):
    has_digit = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

pwd = input("Enter password: ")
print(check_password(pwd))
```
