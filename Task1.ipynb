```python
#2
# List of student marks
marks = [45, 78, 90, 33, 60]

pass_count= 0
fail= 0

for score in marks:
    if score >= 50:
        pass_count=pass_count+1  # Increment pass_count
    else:
        fail=fail+1  # Increment fail

# Printing final result
print("Total Students:", len(marks))
print("Number of Pass Students:", pass_count)
print("Number of Fail Students:", fail)
```

```python
#3
# Original list of names
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []
for name in names:
    #remove extra spaces and convert to lowercase
    cleaned_name = name.strip().lower()
    # Add cleaned name to the new list
    cleaned_names.append(cleaned_name)

# Print the cleaned list
print("Cleaned Names List:", cleaned_names)
```

```python
#4
# List of messages to analyze
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    msg_length = len(msg)

    # Print message and its length
    print("Message:", msg)
    print("Length:", msg_length)

    # Check if message length is greater than 10
    if msg_length > 10:
        print("Status: Long message")
    else:
        print("Status: Short message")
```

```python
#5
# List of system log messages
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0
for log in logs:
    # Check if the log entry is an ERROR
    if log == "ERROR":
        error_count += 1
# Print the total number of errors
print("Total ERROR messages:", error_count)
```
