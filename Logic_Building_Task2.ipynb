```python
#1
sentence="python is easy and python is powerful"
words=sentence.split()
unique_words=set(words)
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
#2
# Employee salary details
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
# Find the employee with the highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print(f"Highest Salary: {highest_employee} - {highest_salary}")
```

```python
#3
numbers = [45, 22, 89, 10, 66]

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print(f"List: {numbers}")
print(f"Max: {maximum}")
print(f"Min: {minimum}")
```

```python
#4
prices = [450, 1200, 899, 1500, 300]

count = 0
for price in prices:
    if price > 1000:
        count += 1

print(f"Products above 1000: {count}")
```

```python
#5
attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
total_days = len(attendance)
attendance_percentage = (present_count / total_days) * 100

print(f"Attendance Percentage: {attendance_percentage}")
```

```python
#6
phone_numbers = [9876543210, 9123456789, 9876543210]

unique_phone_numbers = set(phone_numbers)

print(f"Unique phone numbers: {unique_phone_numbers}")
```

```python
#7
text="python"

char_count={}
for char in text:
    if char in char_count:
        char_count[char]+=1
    else:
        char_count[char]=1

print(char_count)
```
